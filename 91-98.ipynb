91. 如何从一个常规数组中创建记录数组(record array)

```python
import numpy as np
z=np.array([('hello',2.5,3),
           ('world',3.5,2)])
#从(flat)数组列表创建记录数组参数：arrayList：list或tuple
类数组对象的列表（例如，列表，元组和
R=np.core.records.fromarrays(z.T,
                            names='col1,col2,col3',
                            formats='S8,f8,i8')
print(R)
```

94. 思考一个10x3的矩阵，如何分解出有不全相同值的行

```python
z=np.random.randint(0,5,(10,3))
print(z)
e=np.all(z[:,1:]==z[:,:-1],axis=1)#二维切片的使用，前面是行，后面是列
u=z[~e]
print(u)
u=z[z.max(axis=1)!=z.min(axis=1),:]# axis=0 代表行 , axis=1 代表列
print(u)
```

95. 将一个整数向量转换为二进制矩阵（二进制矩阵，除了0就是1）

```python
import numpy as np
i=np.array([0,1,2,3,15,16,32,64,128])
print(i)
#.astype(bool),astype的中文意思，作为布尔类型，也就是true or false
b=((i.reshape(-1,1)&(2**np.arange(8)))!=0).astype(int)#.astype(int)是把布尔类型的转换为为整型数据（0或1）
print(b)
print(b[:,::-1])#行全取完，列倒着取
```

96. 给定一个二维数组，如何提取出唯一的行?

```python
z=np.random.randint(0,2,(6,3))
print(z)
t=np.ascontiguousarray(z).view(np.dtype((np.void,z.dtype.itemsize*z.shape[1])))
_,idx=np.unique(t,return_index=True)
uz=z[idx]
print(uz)
```

98. 考虑一个由两个向量描述的路径(X,Y)，如何用等距样例(equidistant samples)对其进行采样(sample)

```python
phi=np.arange(0,10*np.pi,0.1)
print(phi)
a=1
x=a*phi*np.cos(phi)
y=a*phi*np.sin(phi)
dr=(np.diff(x)**2+np.diff(y)**2)**.5
r=np.zeros_like(x)
r[1:]=np.cumsum(dr)
r_int=np.linspace(0,r.max(),200)
x_int=np.interp(r_int,r,x)
y_int=np.interp(r_int,r,y)
```
